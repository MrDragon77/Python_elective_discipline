Задание 3.

```python
print("Загадайте число (!)")
print("Введите интервал.")
min_num = int(input("минимум: "))
max_num = int(input("максимум: "))
if max_num <= min_num:
    print("Ошибка: Максимум должен быть больше Минимума!")
    exit()

while True:
    cur_len = max_num - min_num
    if cur_len > 0:
        middle = ((max_num + min_num) // 2) + 1
        while True:
            ans = input(f"Число меньше {middle}? (y/n): ")
            if ans != "y" and ans != "n":
                print("Ошибка: Неверный ввод!")
                continue
            break
        if ans == "y":
            max_num = middle - 1
        if ans == "n":
            min_num = middle
    else:
        print(f"Загаданное число - {max_num}")
        break

            
        
```

Задание 2.

```python
import random
n = 20
max_num = 100
groups = 10

list = [random.randint(0, max_num) for _ in range(n)]

print(f"Список: {list}")
gistogram = [0 for _ in range (groups)]
print(f"Гистограмма: {gistogram}")
for num in list:
    group = num % groups
    gistogram[group] += 1
print(f"Распределение: {gistogram}")

probability = [num / n for num in gistogram]
print(f"Вероятности: {probability}")

    
```

Задание 3.

```python
import random

n = int(input("Введите размерность векторов: "))
vector1 = [random.randint(0, 100) for _ in range(n)]
vector2 = [random.randint(0, 100) for _ in range(n)]
print(f"Вектор №1: {vector1}")
print(f"Вектор №2: {vector2}")

scalar = int(input("Введите скаляр: "))

sum = []
multiply = []
for i in range(n):
    sum.append(vector1[i] + vector2[i])
    multiply.append(vector1[i] * vector2[i])
print(f"\nПокомпонентная сумма: {sum}")
print(f"Покомпонентное умножение: {multiply}")

#norms
norm1 = 0
norm2 = 0
for i in range(n):
    norm1 += vector1[i] * vector1[i]
    norm2 += vector2[i] * vector2[i]

scalar_multiply = []
if norm1 > norm2:
    print("\nНорма больше у Вектора №1. ", end = "")
    for i in range(n):
        scalar_multiply.append(vector1[i] * scalar)
else:
    print("\nНорма больше у Вектора №2. ", end = "")
    for i in range(n):
        scalar_multiply.append(vector2[i] * scalar)
print(f"Произведение на скаляр для этого вектора: {scalar_multiply}")

```

```python
import random

m = 5
n = 4
matrix = [[random.randint(0, 100) for _ in range(n)] for _ in range(m)]
print("Матрица: ")
for i in range(m):
    print(matrix[i])

vector = [random.randint(0, 100) for _ in range(n)]
print(f"\nВектор: \n{vector}")

result = []
for i in range(m):
    line_res = 0
    for j in range(n): 
        line_res += matrix[i][j] * vector[j]
    result.append(line_res)
print(f"Умножение матрицы на вектор:\n{result}")
```

Задание 5.

```python
import random

n = 10
list = [random.randint(-100, 100) for _ in range(n)]
print(f"Список: {list}")

last_left_positive = 0
left_positive_list = []
for num in list:
    if num >= 0:
        last_left_positive = num
        left_positive_list.append(-1)
    if num < 0:
        left_positive_list.append(last_left_positive)

last_right_positive = 0
new_list = list.copy()
for i, num in enumerate(reversed(list)):
    if num >= 0:
        last_right_positive = num
    if num < 0:
        new_list[n - i - 1] = (left_positive_list[n - i - 1] + last_right_positive) / 2

print(f"\nНовый список: {new_list}")




```

Задание 6.

```python
import random

n = 10
list = [random.randint(-10, 10) for _ in range(n)]
print(f"Список: {list}")

f = 3
filter = [random.randint(-10, 10) for _ in range(f)]
print(f"Фильтр: {filter}")

if f > n:
    print("Ошибка! Фильтр должен быть меньше списка!")
    exit()

filtered_list = []
for i in range(n - f + 1):
    accumulator = 0
    for j in range(f):
        accumulator += list[i + j] * filter[j]
    filtered_list.append(accumulator)

print(f"Отфильтрованный список: {filtered_list}")
```
