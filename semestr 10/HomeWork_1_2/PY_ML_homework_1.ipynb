```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
# разбейте исходную таблицу на 2: по полу пассажиров
# удалите пустые строки или заполните их соответствующими символами
# Построить гистограмму распределения пассажиров по возрасту
# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
# визуализировать возраст при помощи различных настроек метода plot
```

```python
# 1. разбейте исходную таблицу на 2: по полу пассажиров
df_male = df[df['Sex'] == 'male'].copy()
df_female = df[df['Sex'] == 'female'].copy()

print(f"Мужчины: {len(df_male)}")
print(f"Женщины: {len(df_female)}")
```

```python
# 2. удалите пустые строки или заполните их соответствующими символами
print("before: \n")
print("male:")
print(df_male.isnull().sum())
print("female:")
print(df_female.isnull().sum())

df_male = df_male.dropna()
df_female = df_female.dropna()

print("after: \n")
print("male:")
print(df_male.isnull().sum())
print("female:")
print(df_female.isnull().sum())
```

```python
# 3. Построить гистограмму распределения пассажиров по возрасту
df_male['Age'].hist(bins=20, alpha=0.7, label='Мужчины', color='blue')
df_female['Age'].hist(bins=20, alpha=0.7, label='Женщины', color='red')
plt.title('Распределение пассажиров по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 4. Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
print("Мужчины")
print("Мин:", df_male['Age'].min())
print("Макс:", df_male['Age'].max())
print("Среднее:", df_male['Age'].mean())
print("Медиана:", df_male['Age'].median())
print("Мода:", df_male['Age'].mode()[0])

print("\nЖенщины")
print("Мин:", df_female['Age'].min())
print("Макс:", df_female['Age'].max())
print("Среднее:", df_female['Age'].mean())
print("Медиана:", df_female['Age'].median())
print("Мода:", df_female['Age'].mode()[0])
```

```python
# 5. визуализировать возраст при помощи различных настроек метода plot
bins = 20
male_counts, bins = np.histogram(df_male['Age'], bins=bins)
female_counts, _ = np.histogram(df_female['Age'], bins=bins)

plt.barh(bins[:-1], -male_counts, height=np.diff(bins), color='blue', label='Мужчины')
plt.barh(bins[:-1], female_counts, height=np.diff(bins), color='red', alpha=0.7, label='Женщины')
plt.title('Возрастная пирамида')
plt.xlabel('Кол-во пассажиров')
plt.legend()
```
